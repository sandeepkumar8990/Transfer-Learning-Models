# Inception_v3

Inception v3 is a widely-used image recognition model that has been shown to attain greater than 78.1% accuracy on the ImageNet dataset. The model is the culmination of many ideas developed by multiple researchers over the years. It is based on the original paper: "Rethinking the Inception Architecture for Computer Vision" by Szegedy.

The model itself is made up of symmetric and asymmetric building blocks, including convolutions, average pooling, max pooling, concats, dropouts, and fully connected layers. Batchnorm is used extensively throughout the model and applied to activation inputs. Loss is computed via Softmax.


<img src='Picture4.png'/>

# Layers
Inception-v3 is a convolutional neural network that is 48 layers deep.

# Data:
ImageNet Large Scale Visual Recognition Challenge 2012 classification dataset, consisting of 1.2 million training images, with 1,000 classes of objects.

# Input dimensions:
The network requires  an image input size of 299-by-299 (299,299,3)

# No of classes:
The pretrained network can classify images into 1000 object categories, such as keyboard, mouse, pencil, and many animals. 


```python
from keras.applications.inception_v3 import InceptionV3
incep=InceptionV3()
```

```python
print(incep.summary())
```

```python
from keras.preprocessing.image import load_img
image=load_img('C:/Users/sandeep/Desktop/INTERNSHIP/data/keyboard.jpg',target_size=(299,299,3))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(image)
```

```python
from keras.preprocessing.image import img_to_array
# convert the image pixels to a numpy array
keyboard = img_to_array(image)

# reshape data for the model
keyboard = keyboard.reshape((1, keyboard.shape[0], keyboard.shape[1], keyboard.shape[2]))

from keras.applications.inception_v3 import preprocess_input
# prepare the image for the VGG model
keyboard= preprocess_input(keyboard)

from keras.applications.inception_v3 import decode_predictions
# predict the probability across all output classes
yhin = incep.predict(keyboard)
# convert the probabilities to class labels
labin = decode_predictions(yhin)
# retrieve the most likely result, e.g. highest probability
labin = labin[0][0]
# print the classification
print('%s (%.2f%%)' % (labin[1], labin[2]*100))
```

It correctly predicts as keyboard
