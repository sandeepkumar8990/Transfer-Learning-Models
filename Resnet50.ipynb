# Transfer Learning- Resnet50

ResNet, short for Residual Networks is a classic neural network used as a backbone for many computer vision tasks. This model was the winner of ImageNet challenge in 2015.

There are many variants of ResNet architecture i.e. same concept but with a different number of layers. We have ResNet-18, ResNet-34, ResNet-50, ResNet-101, ResNet-110, ResNet-152, ResNet-164, ResNet-1202 etc.

# Architecture
Now we’ll talk about the architecture of ResNet50. The architecture of ResNet50 has 4 stages as shown in the diagram below.

<img src='Picture3.png'/>

# Input shape:
The model expects color images to have the square shape 224×224.(height,breadth)

# No of classes:
This model recognizes the 1000 different classes of objects in the ImageNet 2012 Large Scale Visual Recognition Challenge. The model consists of a deep convolutional net using the ResNet-50 architecture that was trained on the ImageNet-2012 data set.

# Loss function:
all layers have been trained for 200 epochs with categorical cross-entropy cost function

# Applications
This architecture can be used on computer vision tasks such as image classififcation, object localisation, object detection.

and this framework can also be applied to non computer vision tasks to give them the benifit of depth and to reduce the computational expense also.


```python
from keras.applications.resnet50 import ResNet50
resnet=ResNet50()
```

```python
print(resnet.summary())
```

```python
from keras.preprocessing.image import load_img,img_to_array
image=load_img('C:/Users/sandeep/Desktop/INTERNSHIP/data/monitor.jpg',target_size=(224,224))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(image)
```

```python
# convert the image pixels to a numpy array
monitor = img_to_array(image)

# reshape data for the model
monitor = monitor.reshape((1, monitor.shape[0], monitor.shape[1], monitor.shape[2]))

from keras.applications.resnet50 import preprocess_input
# prepare the image for the VGG model
monitor= preprocess_input(monitor)

from keras.applications.resnet50 import decode_predictions
# predict the probability across all output classes
yhres = resnet.predict(monitor)
# convert the probabilities to class labels
labres = decode_predictions(yhres)
# retrieve the most likely result, e.g. highest probability
labres = labres[0][0]
# print the classification
print('%s (%.2f%%)' % (labres[1], labres[2]*100))
```

This model correctly detect the as monitor
